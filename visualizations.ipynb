# Neutral Emotion

```python
y_neutral, sr_neutral = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-01-01-01-01-01.wav")
y_calm, sr_calm = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-02-01-01-01-01.wav")
y_happy, sr_happy = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-03-01-01-01-01.wav")
y_sad, sr_sad = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-04-01-01-01-01.wav")
y_angry, sr_angry = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-05-01-01-01-01.wav")
y_fearful, sr_fearful = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-06-01-01-01-01.wav")
y_disgust, sr_disgust = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-07-01-01-01-01.wav")
y_surprised, sr_surprised = librosa.load("D:\\Desktop\\Work\\capstone\\capstone-project_computer-vision\\data_files\\voice_training\\Actor_01\\03-01-08-01-01-01-01.wav")
```

## Calculating STFTs...




```python
D_neutral = librosa.stft(y_neutral)
D_calm = librosa.stft(y_calm)
D_happy = librosa.stft(y_happy)
D_sad = librosa.stft(y_sad)
D_angry = librosa.stft(y_angry)
D_fearful = librosa.stft(y_fearful)
D_disgust = librosa.stft(y_disgust)
D_surprised = librosa.stft(y_surprised)
```

## MFCC

```python
fig, ax = plt.subplots(nrows = 8)
fig.set_figheight(60)
fig.set_figwidth(10)

mfcc = librosa.feature.mfcc(y=y_neutral, sr=sr_neutral, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[0])
ax[0].set(title='Neutral')

mfcc = librosa.feature.mfcc(y=y_calm, sr=sr_calm, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[1])
ax[1].set(title='Calm')

mfcc = librosa.feature.mfcc(y=y_happy, sr=sr_happy, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[2])
ax[2].set(title='Happy')

mfcc = librosa.feature.mfcc(y=y_sad, sr=sr_sad, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[3])
ax[3].set(title='Sad')

mfcc = librosa.feature.mfcc(y=y_angry, sr=sr_angry, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[4])
ax[4].set(title='Angry')

mfcc = librosa.feature.mfcc(y=y_fearful, sr=sr_fearful, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[5])
ax[5].set(title='Fearful')

mfcc = librosa.feature.mfcc(y=y_disgust, sr=sr_disgust, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[6])
ax[6].set(title='Disgust')

mfcc = librosa.feature.mfcc(y=y_surprised, sr=sr_surprised, n_mfcc = 40)
librosa.display.specshow(mfcc, x_axis='time', ax = ax[7])
ax[7].set(title='Surprised')
```

## Mel spectogram

```python
fig, ax = plt.subplots(nrows = 8)
fig.set_figheight(60)
fig.set_figwidth(10)

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_neutral, sr=sr_neutral), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[0])
ax[0].set(title='Neutral')
fig.colorbar(img, ax=ax[0], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_calm, sr=sr_calm), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[1])
ax[1].set(title='Calm')
fig.colorbar(img, ax=ax[1], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_happy, sr=sr_happy), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[2])
ax[2].set(title='Happy')
fig.colorbar(img, ax=ax[2], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_sad, sr=sr_sad), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[3])
ax[3].set(title='Sad')
fig.colorbar(img, ax=ax[3], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_angry, sr=sr_angry), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[4])
ax[4].set(title='Angry')
fig.colorbar(img, ax=ax[4], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_fearful, sr=sr_fearful), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[5])
ax[5].set(title='Fearful')
fig.colorbar(img, ax=ax[5], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_disgust, sr=sr_disgust), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[6])
ax[6].set(title='Disgust')
fig.colorbar(img, ax=ax[6], format="%+2.f dB")

M_db = librosa.power_to_db(librosa.feature.melspectrogram(y=y_surprised, sr=sr_surprised), ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax[7])
ax[7].set(title='Surprised')
fig.colorbar(img, ax=ax[7], format="%+2.f dB")
```

## Spectral Contrast

```python
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
```

```python
plt.imshow(normalize(librosa.feature.spectral_contrast(y = y_neutral, sr = sr_neutral), axis=1), aspect='auto', origin='lower', cmap='coolwarm')
plt.title("Neutral")
```
